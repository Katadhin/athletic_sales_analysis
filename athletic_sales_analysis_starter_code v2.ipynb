```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the athletic_sales_2020.csv and athletic-sales_2021.csv CSV files into DataFrames.
sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')

#List column names for each DataFrame
print(sales_2020_df.columns)
print(sales_2021_df.columns)

# describe the 2020 DataFrame
print(sales_2020_df.describe())

# Print column names for the 2021 DataFrame
print(sales_2021_df.columns)


```

```python
# Display the 2020 sales DataFrame
sales_2020_df.head()

```

```python
# Display the 2021 sales DataFrame
sales_2021_df.head()

# Convert the invoice_date column to datetime
sales_2020_df['invoice_date'] = pd.to_datetime(sales_2020_df['invoice_date'])

#Identify null values in the dataframes
print(sales_2020_df.isnull().sum())
print(sales_2021_df.isnull().sum())


```

```python
# Check the 2020 sales data types.
sales_2020_df.dtypes

# Check the 2021 sales data types.
sales_2021_df.dtypes


```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_df = pd.concat([sales_2020_df, sales_2021_df], ignore_index=True)

# Display the combined DataFrame
combined_sales_df.head()


```

```python
# Check if any values are null.
combined_sales_df.isnull().sum()

```

```python
# Check the data type of each column
combined_sales_df.dtypes

```

```python
# Convert the "invoice_date" to a datetime datatype

sales_2020_df['invoice_date'] = pd.to_datetime(sales_2020_df['invoice_date'], format='%m/%d/%y')
sales_2021_df['invoice_date'] = pd.to_datetime(sales_2021_df['invoice_date'], format='%m/%d/%y')

print(sales_2020_df.dtypes)
print(sales_2021_df.dtypes)

print(sales_2020_df.head())
print(sales_2021_df.head())



```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Group by region, state, and city, then sum the units_sold
region_sales_df = combined_sales_df.groupby('region')['units_sold'].sum().reset_index().rename(columns={'units_sold': 'Total_Products_Sold'}).sort_values('Total_Products_Sold', ascending=False).head(5)
state_sales_df = combined_sales_df.groupby('state')['units_sold'].sum().reset_index().rename(columns={'units_sold': 'Total_Products_Sold'}).sort_values('Total_Products_Sold', ascending=False).head(5)
city_sales_df = combined_sales_df.groupby('city')['units_sold'].sum().reset_index().rename(columns={'units_sold': 'Total_Products_Sold'}).sort_values('Total_Products_Sold', ascending=False).head(5)

# Display the top 5 results for each category
print("Top 5 Regions by Total Products Sold:")
print(region_sales_df)

print("\nTop 5 States by Total Products Sold:")
print(state_sales_df)

print("\nTop 5 Cities by Total Products Sold:")
print(city_sales_df)

                      
```

#### Using `pivot_table`

```python
# Create a pivot table for regions
region_sales_pivot = combined_sales_df.pivot_table(index='region', values='units_sold', aggfunc='sum').reset_index().rename(columns={'units_sold': 'Total_Products_Sold'}).sort_values('Total_Products_Sold', ascending=False).head(5)

# Create a pivot table for states
state_sales_pivot = combined_sales_df.pivot_table(index='state', values='units_sold', aggfunc='sum').reset_index().rename(columns={'units_sold': 'Total_Products_Sold'}).sort_values('Total_Products_Sold', ascending=False).head(5)

# Create a pivot table for cities
city_sales_pivot = combined_sales_df.pivot_table(index='city', values='units_sold', aggfunc='sum').reset_index().rename(columns={'units_sold': 'Total_Products_Sold'}).sort_values('Total_Products_Sold', ascending=False).head(5)

# Display the top 5 results for each category
print("Top 5 Regions by Total Products Sold using Pivot Table:")
print(region_sales_pivot)

print("\nTop 5 States by Total Products Sold using Pivot Table:")
print(state_sales_pivot)

print("\nTop 5 Cities by Total Products Sold using Pivot Table:")
print(city_sales_pivot)


```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Ensure invoice_date is in datetime format and set as index
womens_athletic_footwear_sales = womens_athletic_footwear_sales.set_index(pd.to_datetime(womens_athletic_footwear_sales['invoice_date']))

# Now, resample by week and sum the total sales
weekly_womens_footwear_sales = womens_athletic_footwear_sales.resample('W')['total_sales'].sum().reset_index().rename(columns={'total_sales': 'Total_Sales'}).sort_values('Total_Sales', ascending=False)

# Display the top weeks for women's athletic footwear sales
print("Weeks with the Highest Sales for Women's Athletic Footwear:")
print(weekly_womens_footwear_sales.head())


```
